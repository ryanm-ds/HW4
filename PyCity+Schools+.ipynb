```python
for school in schools:
    
    
    loopdf = schoolinfo_df[schoolinfo_df['school_name']==school]
    school_name = loopdf['school_name'].iloc[0]
    totalstudents = loopdf['Student ID'].count()
    totalbudget = loopdf['budget'].iloc[0]
    perstudent = totalbudget/totalstudents
    avgmath = loopdf['math_score'].mean()
    avgreading = loopdf['reading_score'].mean()
    pct_math = ((loopdf[loopdf['math_score']>=70].count())/totalstudents)[0]
    pct_reading = ((loopdf[loopdf['reading_score']>=70].count())/totalstudents)[0]
    ovr_avg = (pct_math+pct_reading)/2
    
    
    
    school_summary =school_summary.append({'School':school_name,'Total Students':totalstudents, 
                                            'Total School Budget':totalbudget,'Per Student Budget':perstudent,
                                            'Average Math Score':avgmath, 
                                       'Average Reading Score':avgreading,'% Passing Math':pct_math,
                                            '% Passing Reading':pct_reading,'% Overall Passing Rate':ovr_avg},ignore_index=True)
    
    
```

```python
grades_list = ['9th', '10th', '11th', '12th']
```

```python
## function to get each high school grade and combine them ##
def high_school_grade_splitter(dataframe, grade_list):
    list_grades = []
    for x in grade_list:
        list_grades.append(pd.Series(dataframe[dataframe['grade']==x].groupby(by='school_name')['math_score'].mean(),
                       name=x))
        
    return pd.concat(list_grades, axis=1)
```

```python
combined_math = high_school_grade_splitter(schoolinfo_df, grades_list)
```

```python
combined_math
```

```python
def reading_score_grade(dataframe,grades):
    grade_list = []
    for grade in grades:
        grade_list.append(pd.Series(schoolinfo_df[schoolinfo_df['grade']==grade].groupby('school_name')['reading_score']
                                    .mean(),name=grade))
    return pd.concat(grade_list,axis=1)
```

```python
readingby_grade = reading_score_grade(schoolinfo_df,grades_list)
```
